#### Using Functions
You may want to repeat such calculations on several texts, but it is tedious to keep retyping the formula. Instead, you can come up with your own name for a task, like "lexical_diversity" or "percentage", and associate it with a block of code.

```python
def lexical_diversity(text):
    return len(set(text))/len(text)
```

```python
def percentage(count, total):
    return 100 * count/total
```

```python
lexical_diversity(text3)
```

```python
lexical_diversity(text5)
```

```python
percentage(4,5)
```

```python
percentage(text4.count('a'), len(text4))
```

### 3-1 Frequency Distributions
tells us the frequency of each vocabulary item in the text  
"distribution" because it tells us how the total number of word tokens in the text are distributed across the vocabulary items  
nltk에서 제공하는 FreqDist 사용

FreqDist 함수들  
- fdist1.most_common(#) : #개의 가장 흔하게 나오는 단어들과 그 수 - a list of the # most frequent words in the text
- fdist1.plot(#, cumulative = bool) : 위를 그래프로 그리기, 수가 누적됨
- fdist1.plot(#, cumulative = bool) : 그래프로 그리기, 수가 누적되지 않고 말 그대로 그 수가 표현된 그래프
- fdist1.hapaxes() : 한 번만 나오는 단어들 - infrequent words, unique

```python
fdist1 = FreqDist(text1)
```

```python
print(fdist1)
```

```python
fdist1.most_common(50)
```

```python
fdist1.plot(50, cumulative = True)
```

```python
fdist1.plot(50, cumulative = False)
```

```python
fdist1.hapaxes()
```

## 3.2 Fine-grained Selection of Words

```python
long_words = [w for w in V if len(w) > 15]
```

```python
sorted(long_words)
```

```python
long_words2 = [word for word in V if len(word) > 15]
sorted(long_words2)
```

```python
fdist5 = FreqDist(text5)
```

```python
sorted(w for w in set(text5) if len(w) > 7 and fdist5[w] > 7)
```

Notice how we have used two conditions: len(w) > 7 ensures that the words are longer than seven letters, and fdist5[w] > 7 ensures that these words occur more than seven times.

## 3-4 Counting Other Things
Counting words is useful, but we can count other things too.  
We can also count length of each word and make it into a list, using FreqDist  
여기서는 우리가 단어 길이의 빈도수를 센다.

- fdist : = FreqDist(len(w) for w in text1)일 때, 각 단어 길이가 얼마나 자주 나오는지를 찾아서 dictionary로 표현
- fdist.plot() : 위를 그래프로 표현
- fdist.most_common() : 각 단어 길이가 얼마나 자주 나오는지를 tuple 묶음으로 이루어진 list로 표현
- fdist.max() : 가장 많이 나온 단어 길이
- fdist[#] : 단어 길이 #가 얼마나 자주 나왔는지
- fdist.freq(#) : 얼마나 #이 자주 나오는지 percentage가 아닌 소수인 확률로 표현 (100을 곱하면 percentage가 됨!)

##### sorted(w for w in set(text1) if condition)  

함수들 (위 condition 부분, if 이후에 나옴)  
- 단어리스트.endswith(단어) : 단어리스트에서 단어로 끝나는 것들 찾아줌 - s.endswith(t) : test if s ends with t
- '알파벳' in 단어리스트 : 단어리스트 중 알파벳이 들어가는 것들을 찾아줌 - t in s :	test if t is a substring of s
- 단어리스트.istitle() : 단어리스트에서 맨 앞 글자는 대문자, 나머지는 소문자인 것들 찾아줌 - s.istitle() : test if s contains cased characters and is titlecased (i.e. all words in s have initial capitals)
- 단어리스트.isdigit() : 단어리스트에서 숫자인 것들 찾아줌 - s.isdigit() : test if s is non-empty and all characters in s are digits

```python
sorted(w for w in set(text1) if w.endswith('ableness'))
```

```python
sorted(term for term in set(text4) if 'gnt' in term)
```

```python
sorted(item for item in set(text6) if item.istitle())
```

```python
sorted(item for item in set(sent7) if item.isdigit())
```

We can also create more complex conditions.  
If we have two conditions c1 and c2, then we can combine them to form a new condition: c1 and c2, c1 or c2.

```python
sorted(w for w in set(text7) if '-' in w and 'index' in w)
```

```python
sorted(wd for wd in set(text3) if wd.istitle() and len(wd) > 10)
```

```python
sorted(w for w in set(sent7) if not w.islower())
```

```python
sorted(t for t in set(text2) if 'cie' in t or 'cei' in t)
```
